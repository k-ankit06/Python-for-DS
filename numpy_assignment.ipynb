# Assignment 1: Working with NumPy

Objective: Understand the basics of NumPy, array creation, and manipulation.

```
import numpy as np
```

## 1. 1D NumPy Array Operations

```
arr1 = np.arange(1, 21)
arr1
```

```
np.sum(arr1), np.mean(arr1), np.median(arr1), np.std(arr1)
```

```
np.where(arr1 > 10)
```

## 2. 2D NumPy Array

```
arr2 = np.arange(1, 17).reshape(4, 4)
arr2
```

```
arr2.T
```

```
np.sum(arr2, axis=1), np.sum(arr2, axis=0)
```

## 3. Random 3x3 Arrays

```
a = np.random.randint(1, 21, (3, 3))
b = np.random.randint(1, 21, (3, 3))
a, b
```

```
a+b, a-b, a*b
```

```
np.dot(a, b)
```

## 4. Reshape and Slice

```
arr = np.arange(1, 13)
reshaped = arr.reshape(3, 4)
reshaped[:2, -2:]
```
