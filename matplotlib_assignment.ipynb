# Assignment 3: Working with Matplotlib

Objective: Practice data visualization.

```
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot

```
x = [1,2,3,4,5]
y = [10,15,25,30,50]
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

## Bar Graph

```
students = ['John','Jane','Alice','Bob']
marks = [75,85,60,90]
plt.bar(students, marks)
plt.title('Student Marks')
plt.show()
```

## Pie Chart

```
regions = ['North America','Europe','Asia','Others']
revenue = [45,25,20,10]
plt.pie(revenue, labels=regions, autopct='%1.1f%%', explode=(0.1,0,0,0))
plt.title('Revenue Distribution')
plt.show()
```

## Histogram

```
data = np.random.randint(1, 101, 1000)
plt.hist(data, bins=20)
plt.title('Histogram')
plt.show()
```
